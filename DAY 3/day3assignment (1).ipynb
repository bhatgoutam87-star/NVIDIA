```python
# Intel Image Classification using TensorFlow

import os
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"   # Hide TensorFlow warnings

import tensorflow as tf
import zipfile

# Extract dataset
with zipfile.ZipFile("archive.zip", "r") as zip_ref:
    zip_ref.extractall("intel_dataset")

# Dataset paths
train_path = "./intel_dataset/Intel-Image-Classification/train"
val_path = "./intel_dataset/Intel-Image-Classification/val"
test_path = "./intel_dataset/Intel-Image-Classification/test"

# Load datasets
train_ds = tf.keras.utils.image_dataset_from_directory(
    train_path,
    image_size=(150, 150),
    batch_size=32
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    val_path,
    image_size=(150, 150),
    batch_size=32
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    test_path,
    image_size=(150, 150),
    batch_size=32
)

# Build CNN Model
model = tf.keras.Sequential([
    tf.keras.Input(shape=(150, 150, 3)),
    tf.keras.layers.Rescaling(1./255),

    tf.keras.layers.Conv2D(32, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(64, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(128, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),

    tf.keras.layers.Dense(6, activation='softmax')
])

# Compile
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5
)

# Evaluate
loss, accuracy = model.evaluate(test_ds)

print("\nTest Accuracy:", round(accuracy * 100, 2), "%")
```
