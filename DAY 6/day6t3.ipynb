````python
from google import genai
from IPython.display import display, HTML

# Initialize the client with your working API key
client = genai.Client(api_key="API_KEY")

prompt = """
Write a complete, single block of raw HTML/SVG code that renders an ultra-detailed futuristic cyberpunk sports car. Give it glowing purple and cyan neon underglow, aerodynamic body panels, illuminated wheel rims, holographic dashboard projections visible through tinted glass, and reflective chrome accents. Place the vehicle on a dark futuristic city street with subtle rain reflections, digital billboards, and a neon-lit skyline in the background. Use advanced SVG gradients, filters, shadows, and glow effects to create a highly realistic, high-tech appearance. Return ONLY the raw HTML/SVG code, with no markdown formatting, explanations, or additional text.

"""

print("Generating vector graphics... please wait a moment...")

try:
    # Use your working Gemini model to generate the vector code
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=prompt,
    )
    
    # Clean up any potential markdown formatting from the response string
    clean_html = response.text.replace("```html", "").replace("```", "").strip()
    
    # Render the vector artwork cleanly inside your notebook layout
    display(HTML(clean_html))

except Exception as e:
    print(f"\nExecution error: {e}")
````
