```python
import torch
import time
import psutil

from transformers import (
    pipeline,
    AutoTokenizer,
    AutoModel,
    AutoModelForCausalLM
)

print("="*60)
print("HUGGING FACE AI WORKSTATION")
print("="*60)

# ---------------------------------------------------
# Device Information
# ---------------------------------------------------

device = "cuda" if torch.cuda.is_available() else "cpu"

print("\nDevice:", device)

if torch.cuda.is_available():

    print(
        "GPU:",
        torch.cuda.get_device_name(0)
    )

# ---------------------------------------------------
# Tokenizer Demo
# ---------------------------------------------------

print("\nTOKENIZATION")

tokenizer_gpt2 = AutoTokenizer.from_pretrained(
    "distilgpt2"
)

text = """
Artificial Intelligence is transforming
education and cybersecurity.
"""

tokens = tokenizer_gpt2.tokenize(text)

print("Tokens:")
print(tokens)

print("Token Count:", len(tokens))

# ---------------------------------------------------
# Embedding Demo
# ---------------------------------------------------

print("\nEMBEDDINGS")

# Use a BERT tokenizer for the BERT model
tokenizer_bert = AutoTokenizer.from_pretrained(
    "bert-base-uncased"
)

model = AutoModel.from_pretrained(
    "bert-base-uncased"
)

inputs = tokenizer_bert(
    text,
    return_tensors="pt"
)

with torch.no_grad():

    outputs = model(**inputs)

embedding = outputs.last_hidden_state

print(
    "Embedding Shape:",
    embedding.shape
)

# ---------------------------------------------------
# Text Generation
# ---------------------------------------------------

print("\nTEXT GENERATION")

generator = pipeline(
    "text-generation",
    model="distilgpt2"
)

result = generator(
    "Machine Learning is",
    max_new_tokens=50
)

print(
    result[0]["generated_text"]
)

# ---------------------------------------------------
# Sentiment Analysis
# ---------------------------------------------------

print("\nSENTIMENT ANALYSIS")

sentiment = pipeline(
    "sentiment-analysis"
)

reviews = [
    "This workshop is amazing.",
    "The session was boring."
]

for review in reviews:

    print(
        sentiment(review)
    )

# ---------------------------------------------------
# Translation
# ---------------------------------------------------

print("\nTRANSLATION")

translator = pipeline(
    "translation_en_to_fr",
    model="t5-small"
)

translated = translator(
    "Artificial Intelligence is transforming education."
)

print(
    translated[0]["translation_text"]
)

# ---------------------------------------------------
# Throughput Benchmark
# ---------------------------------------------------

print("\nBENCHMARK")

prompt = """
Explain Transformers in AI.
"""

start = time.time()

output = generator(
    prompt,
    max_new_tokens=100
)

end = time.time()

generated_text = output[0]["generated_text"]

tokens_generated = len(
    generated_text.split()
)

throughput = (
    tokens_generated /
    (end-start)
)

print(
    "Time:",
    round(end-start,2),
    "sec"
)

print(
    "Tokens/sec:",
    round(throughput,2)
)

# -------------------------------------------------
# Memory Monitoring
# -------------------------------------------------

print("\nMEMORY")

process = psutil.Process()

memory = (
    process.memory_info().rss /
    1024**2
)

print(
    "RAM Usage:",
    round(memory,2),
    "MB"
)

# ---------------------------------------------------
# GPU Monitoring
# ---------------------------------------------------

if torch.cuda.is_available():

    gpu_memory = (
        torch.cuda.memory_allocated()
        /1024**3
    )

    print(
        "GPU Memory:",
        round(gpu_memory,2),
        "GB"
    )

print("\nExecution Complete")
```
