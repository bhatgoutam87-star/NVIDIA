```python
import os
import zipfile
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical # Added for one-hot encoding

# ==========================================
# Removed: Extract ZIP File and Find Dataset Folder
# As tf.keras.datasets.fashion_mnist.load_data() handles data loading directly.
# ==========================================

# ==========================================
# Load Fashion MNIST Dataset
# ==========================================

print("Loading Fashion MNIST dataset...")
# Load Fashion MNIST data
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.fashion_mnist.load_data()
print("Fashion MNIST dataset loaded.")

# ==========================================
# Preprocessing
# ==========================================

img_size = (128, 128) # Target size for the model, upscaled from 28x28
batch_size = 32

print(f"Resizing images to {img_size} and normalizing...")
# Resize images to target_size (e.g., 128x128) and add channel dimension (1 for grayscale)
# tf.image.resize expects float32, so cast first.
train_images = tf.image.resize(tf.cast(train_images[..., tf.newaxis], tf.float32), img_size).numpy()
test_images = tf.image.resize(tf.cast(test_images[..., tf.newaxis], tf.float32), img_size).numpy()

# Normalize pixel values to [0, 1]
train_images = train_images / 255.0
test_images = test_images / 255.0

# Determine number of classes
num_classes = len(np.unique(np.argmax(train_labels, axis=-1) if train_labels.ndim > 1 else train_labels))
# Re-calculating num_classes based on raw labels in case they weren't one-hot encoded yet
# For fashion_mnist, it's 10 classes
if num_classes != 10: # Safety check, it should be 10 for Fashion MNIST
    num_classes = 10

print("One-hot encoding labels...")
# One-hot encode labels
train_labels = to_categorical(train_labels, num_classes=num_classes)
test_labels = to_categorical(test_labels, num_classes=num_classes)

print("Train images shape:", train_images.shape)
print("Train labels shape:", train_labels.shape)
print("Number of classes:", num_classes)

# ==========================================
# Data Generator (for augmentation and validation split)
# ==========================================

print("Setting up ImageDataGenerator...")
datagen = ImageDataGenerator(
    validation_split=0.2, # Use 20% of training data for validation
    # Add common augmentation techniques if desired, e.g.:
    # rotation_range=10,
    # zoom_range=0.1,
    # width_shift_range=0.1,
    # height_shift_range=0.1,
    # horizontal_flip=True,
)

# Using flow from arrays for train and validation split
train_generator = datagen.flow(
    train_images, train_labels,
    batch_size=batch_size,
    subset='training'
)

val_generator = datagen.flow(
    train_images, train_labels,
    batch_size=batch_size,
    subset='validation'
)
print("Data generators created.")

# ==========================================
# CNN Model
# ==========================================

print("Building CNN model...")
model = Sequential([
    Conv2D(32, (3,3), activation='relu',
           input_shape=(img_size[0], img_size[1], 1)), # Changed input_shape to 1 channel for grayscale
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),

    Dense(256, activation='relu'),
    Dropout(0.5),

    Dense(num_classes, activation='softmax') # num_classes will now be 10
])

model.compile(
    optimizer=Adam(1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ==========================================
# Training
# ==========================================

print("Starting model training...")
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10
)
print("Model training finished.")

# ==========================================
# Accuracy Graph
# ==========================================

print("Plotting accuracy graph...")
plt.figure(figsize=(10,5))

plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')

plt.title("Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
print("Accuracy graph displayed.")

# ==========================================
# Loss Graph
# ==========================================

print("Plotting loss graph...")
plt.figure(figsize=(10,5))

plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')

plt.title("Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
print("Loss graph displayed.")

# ==========================================
# Final Accuracy
# ==========================================

print("Evaluating model on test data...")
loss, acc = model.evaluate(test_images, test_labels) # Evaluate on the independent test set

print(f"\nTest Accuracy: {acc*100:.2f}%") # Changed to Test Accuracy
print("Model evaluation complete.")

# ==========================================
# Save Model
# ==========================================

print("Saving model...")
model.save("cnn_model.h5")

print("Model Saved Successfully!")
```
